```python
from datetime import timedelta
import matplotlib.pyplot as plt
import seaborn as sns

# Make sure date is in datetime format
listing_summary['date'] = pd.to_datetime(listing_summary['date'], errors='coerce')
latest_date = listing_summary['date'].max()
one_year_ago = latest_date - timedelta(days=365)
recent_data = listing_summary[listing_summary['date'] >= one_year_ago].copy()

# Plot: Price Distribution for the last 12 months
plt.figure(figsize=(8, 5))
sns.histplot(data=recent_data, x='price', bins=1000, color='teal')
plt.title('Price Distribution (Last 12 Months)')
plt.xlabel('Price (€ per night)')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

**Identifying Outliers**

```python
Q1 = listing_summary['price'].quantile(0.25)
Q3 = listing_summary['price'].quantile(0.75)
IQR = Q3 - Q1

# Define outlier bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
outliers = listing_summary[(listing_summary['price'] < lower_bound) | (listing_summary['price'] > upper_bound)]
print(f"Lower Bound: {lower_bound:.2f}")
print(f"Upper Bound: {upper_bound:.2f}")
print(f"Number of outliers: {outliers.shape[0]}")
print(f"Total listings: {listing_summary.shape[0]}")
print(f"Percentage of outliers: {(outliers.shape[0] / listing_summary.shape[0]) * 100:.2f}%")
```

```python
# Optional: Preview outliers
outliers[['id', 'price', 'neighbourhood', 'room_type']].head()
```

**Attention:**
Filtered outlier pricing range to nothing greater than €850 to reduce skew and improve visualization accuracy

```python
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import timedelta

# Ensure 'date' column is datetime
airbnb_summary_capped['date'] = pd.to_datetime(airbnb_summary_capped['date'], errors='coerce')

# Filter for last 12 months
latest_date = airbnb_summary_capped['date'].max()
one_year_ago = latest_date - timedelta(days=365)
recent_data = airbnb_summary_capped[airbnb_summary_capped['date'] >= one_year_ago].copy()

# Extract month and group by it
recent_data['month'] = recent_data['date'].dt.to_period('M')
monthly_trend = (
    recent_data
    .groupby('month')['price_capped']
    .mean()
    .reset_index()
)
monthly_trend['month'] = monthly_trend['month'].astype(str)

# Plot the seasonal trend (clean version)
plt.figure(figsize=(10, 5))
sns.lineplot(data=monthly_trend, x='month', y='price_capped', marker='o', color='teal')
plt.title('Average Price per Month (Last 12 Months)')
plt.xlabel('Month')
plt.ylabel('Average Nightly Price (€)')
plt.xticks(rotation=45)
plt.grid(False)               # Remove gridlines
plt.legend([],[], frameon=False)  # Remove legend
plt.tight_layout()
plt.show()
```

```python
# Print seasonal summary
print("\nSeasonal Price Summary (Last 12 Months):")
print("Highest Avg. Price Month:", monthly_trend.loc[monthly_trend['price_capped'].idxmax()].to_dict())
print("Lowest Avg. Price Month:", monthly_trend.loc[monthly_trend['price_capped'].idxmin()].to_dict())
```

**Insight:**

The highest average nightly price occurred in March 2024 (€163), while the lowest was in October 2024 (€109). This suggests that spring may offer stronger earning potential than early autumn, highlighting an opportunity for seasonal pricing adjustments to align with market demand.
